```python
import duckdb
import pandas as pd
```

```python
csv_path = "files/ads_spend_modified.csv"

df = pd.read_csv(csv_path)

con = duckdb.connect("files/ads.duckdb")

df
```

```python
con.execute("""
    CREATE TABLE IF NOT EXISTS ads_data AS SELECT * FROM df LIMIT 0
""")

con.execute("INSERT INTO ads_data SELECT * FROM df")

count = con.execute("SELECT COUNT(*) FROM ads_data").fetchone()[0]
print(f"Registros en ads_data: {count}")

con.close()
```
